```python
import os
import cv2
import math
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.getcwd()
```

```python
files = glob.glob("./Images/boat*.jpg")
files.sort()

images = []
print(len(images))
for filename in files:
    img = cv2.imread(filename)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    images.append(img)

plt.figure(figsize=[30,10])
cols = 3
rows = math.ceil(len(images)/cols)
for i in range(0, len(images)):
    plt.subplot(rows, cols, i+1)
    plt.axis('off')
    plt.imshow(images[i])
plt.show()
```

```python
stitcher = cv2.Stitcher_create()
status, final_img = stitcher.stitch(images)

if status == 0:
    plt.figure(figsize=[30,10])
    plt.imshow(final_img)
plt.show()
```
